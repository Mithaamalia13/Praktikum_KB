# **DATA CLEANING**

### **MENANGANI NILAI NULL**

```python
#mengecek data null
data.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer_string = SimpleImputer(strategy="most_frequent") #modus
```

```python
data['Gender'] = imputer_string.fit_transform(data[["Gender"]])
data['Married'] = imputer_string.fit_transform(data[["Married"]])
data['Dependents'] = imputer_string.fit_transform(data[["Dependents"]])
data['Self_Employed'] = imputer_string.fit_transform(data[["Self_Employed"]])
```

```python
data.isna().sum()
```

```python
imputer_float = SimpleImputer(strategy="mean")
```

```python
data['LoanAmount'] = imputer_float.fit_transform(data[["LoanAmount"]])
data['Loan_Amount_Term'] = imputer_float.fit_transform(data[["Loan_Amount_Term"]])
data['Credit_History'] = imputer_float.fit_transform(data[["Credit_History"]])
```

```python
data.isna().sum()
```

### **MENANGANI RECORD DUPLICATE**
